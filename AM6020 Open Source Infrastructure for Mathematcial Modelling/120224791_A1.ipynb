The [RSA method of encryption](https://simple.wikipedia.org/wiki/RSA_algorithm) is widely used for cyber-security. Each of us probably make use of it many times everyday without thinking about it. Write a function `check_prime(n)` which returns `True` if a number is prime and `False` if this is not the case. 

```python
# Write your code here.
def check_prime(n):
    if n <= 1 or n % 1 > 0 or n==4:
        return False
    for i in range(2, n//2):
        if n % i == 0:
            return False
    return True
```

$\frac{dM}{dt}=V_s\frac{1}{1+(P_N)}-V_m\frac{M}{0.5+M}$  
$\frac{dP_0}{dt}=0.38M-3.2\frac{P_0}{2+P_0}+1.58\frac{P_1}{2+P_1}$  
$\frac{dP_1}{dt}=3.2\frac{P_0}{2+P_0}-6.58\frac{P_1}{2+P_1}+2.5\frac{P_2}{2+P_2}$  
$\frac{dP_2}{dt}=5\frac{P_1}{2+P_1}-2.5\frac{P_2}{2+P_2}-1.9P_2+1.3P_N$  
$\frac{dP_N}{dt}=1.9P_2-1.3P_N$


```python
import math 
import numpy as np
import matplotlib.pyplot as plt

def CircaOsc(Mi, P0i,P1i, P2i, PNi, Vs, Vm):
    Mdot=(Vs*(1/(1+(PNi**4)))-(Vm*(Mi/(0.5+Mi))))
    P0dot=(0.38*Mi)-(3.2*(P0i/(2+P0i)))+1.58*(P1i/(2+P1i))
    P1dot=(3.2*(P0i/(2+P0i)))-(6.58*(P1i/(2+P1i)))+(2.5*(P2i/(2+P2i)))
    P2dot=(5*(P1i/(2+P1i)))-(2.5*(P2i/(2+P2i)))-(1.9*P2i)+(1.3*PNi)
    PNdot=(1.9*P2i)-(1.3*PNi)
    return (Mdot,P0dot, P1dot, P2dot,PNdot)

def euler(h,a,b,M0,P00,P10,P20,PN0,Vs,Vm):
    '''Function for Euler's method'''
    #Initial conditions
    N=int((b-a)/h)
    t=np.linspace(a,b,N+1)
    M=np.zeros((N+1))
    P0=np.zeros((N+1))
    P1=np.zeros((N+1)) 
    P2=np.zeros((N+1)) 
    PN=np.zeros((N+1)) 
    M[0],P0[0],P1[0],P2[0],PN[0]=(M0,P00,P10,P20,PN0)
    # Loop around i=0,1,...,N, remembering that range() only takes integers as arguments
    for i in range(N):
        Mdot,P0dot,P1dot,P2dot,PNdot= CircaOsc(M[i],P0[i],P1[i],P2[i],PN[i],Vs,Vm)
        M[i+1]=M[i]+h*Mdot
        P0[i+1]=P0[i]+h*P0dot
        P1[i+1]=P1[i]+h*P1dot
        P2[i+1]=P2[i]+h*P2dot
        PN[i+1]=PN[i]+h*PNdot
    return (t,M,P0,P1,P2,PN)

euler(0.01,0.0001,150,0.68,0.68,0.68,0.68,0.68,0.76, 0.65)

#Picking 0.01 of the value of t makes the simulation more accurate. It gives the graph more plots and makes it 
#accurate as a model.
```
